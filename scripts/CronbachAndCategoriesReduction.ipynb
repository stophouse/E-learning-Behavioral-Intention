```python
import pandas as pd
df = pd.read_csv('zdalna6.csv')
```

```python
pu_cols = ['PU1','PU2','PU3','PU4','PU5','PU6']
peou_cols = ['PEOU1','PEOU2','PEOU3','PEOU4','PEOU5','PEOU6','PEOU7','PEOU8','PEOU9']
bi_cols = ['BI1','BI2','BI3']
eng_cols = ['ENG1','ENG2']
neg_cols = ['NEG1','NEG2','NEG3','NEG4','NEG5','NEG6','NEG7']
pos_cols = ['POS1','POS2','POS3','POS4']

# Funkcja do Cronbacha α
def cronbach_alpha(items):
    items = items.dropna()
    k = items.shape[1]
    variances = items.var(axis=0, ddof=1)
    total_var = items.sum(axis=1).var(ddof=1)
    alpha = k/(k-1)*(1 - variances.sum()/total_var)
    return alpha

print("PU α:", cronbach_alpha(df[pu_cols]))
print("PEOU α:", cronbach_alpha(df[peou_cols]))
print("BI α:", cronbach_alpha(df[bi_cols]))
print("ENG α:", cronbach_alpha(df[eng_cols]))
print("NEG α:", cronbach_alpha(df[neg_cols]))
print("POS α:", cronbach_alpha(df[pos_cols]))
```

```python
df['PU'] = df[pu_cols].mean(axis=1)
df['PEOU'] = df[peou_cols].mean(axis=1)
df['BI'] = df[bi_cols].mean(axis=1)
df['ENG'] = df[eng_cols].mean(axis=1)
df['NEG'] = df[neg_cols].mean(axis=1)
df['POS'] = df[pos_cols].mean(axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Lista latentnych zmiennych
latent_cols = ['PU', 'PEOU', 'BI', 'ENG', 'NEG', 'POS']

# Obliczenie macierzy korelacji
corr_matrix = df[latent_cols].corr()

# Wyświetlenie mapy cieplnej
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Macierz korelacji zmiennych latentnych")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('zdalna7.csv')
```

```python
social_cols = ['goal_partner', 'goal_social_life', 'goal_move_out']
instrumental_cols = ['goal_diploma', 'goal_skills']

df['goal_social'] = df[social_cols].sum(axis=1)
df['goal_instrumental'] = df[instrumental_cols].sum(axis=1)
```

```python
it_cols = [
    'study program_Computer Science',
    'study program_Cybersecurity'
]

business_cols = [
    'study program_Finance and Accounting',
    'study program_Management'
]

social_cols = [
    'study program_Political Science',
    'study program_Public Administration',
    'study program_Security and International Relations'
]

humanities_cols = [
    'study program_English Philology',
    'study program_German Philology',
    'study program_New Media and Public Relations'
]

applied_cols = [
    'study program_Psychology',
    'study program_Dietetics',
    'study program_Cosmetology',
    'study program_Fashion Design'
]

df['study_program_it'] = df[it_cols].max(axis=1)
df['study_program_business'] = df[business_cols].max(axis=1)
df['study_program_social_sciences'] = df[social_cols].max(axis=1)
df['study_program_humanities'] = df[humanities_cols].max(axis=1)
df['study_program_applied'] = df[applied_cols].max(axis=1)
```
